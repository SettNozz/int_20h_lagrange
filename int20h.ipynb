```python
import sys
import pandas as pd
```

```python
data = pd.read_csv('data_hackathon_v04.csv')
data
```

```python
columns_name = data.columns[0].split('|')
columns_name
```

```python
values = []
for d in range(data.shape[0]):
    item = data.loc[d].values[0]
    value = item.split('|')    
    values.append(value)
```

```python
# print(values[0])
# print(len(values[0]))
```

```python
df = {}
for c in range(12):
    d = []
    df[columns_name[c]] = d
    for value in values:
        d.append(value[c])
        
        
```

```python
df_main = pd.DataFrame(data=df)
```
